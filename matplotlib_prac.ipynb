# **Bar Plot**

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[2,4,6,8]
c=['r','g','b','y']
plt.bar(x,y,color=c)
plt.show()
```

```python
x=["python","C","C++","Java"]
y=[85,70,60,82]
plt.xlabel("Programming Language",fontsize=15)
plt.ylabel("Percentage",fontsize=15)
plt.title("Programming Language Popularity",fontsize=15)
# c=['y','b','m','g']
plt.bar(x,y,width=0.4,color='r',edgecolor='black',label='popularity')
plt.legend()
plt.show()
```

```python
x=["random1","random2","random3","random4"]
y=[85,70,60,82]
z=[20,30,40,50]
plt.xlabel("x label",fontsize=15)
plt.ylabel("y label",fontsize=15)
plt.title("xy bar graph",fontsize=15)
# c=['y','b','m','g']
plt.bar(x,y,width=0.4,color='r',edgecolor='black',label='legend1')
plt.bar(x,z,width=0.4,color='b',edgecolor='black',label='legend2')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=["random1","random2","random3","random4"]
y=[85,70,60,82]
z=[20,30,40,50]
p=np.arange(len(x))
width=0.4
p1=[j+width for j in p]
plt.xlabel("x label",fontsize=15)
plt.ylabel("y label",fontsize=15)
plt.title("xy bar graph",fontsize=15)
# c=['y','b','m','g']

plt.bar(p,y,width,color='r',edgecolor='black',label='legend1')
plt.bar(p1,z,width,color='b',edgecolor='black',label='legend2')

plt.xticks(p+width/2,x,rotation=90)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=["random1","random2","random3","random4"]
y=[85,70,60,82]
z=[20,30,40,50]
p=np.arange(len(x))
width=0.4
p1=[j+width for j in p]
plt.xlabel("y label",fontsize=15)
plt.ylabel("x label",fontsize=15)
plt.title("xy bar graph",fontsize=15)
# c=['y','b','m','g']

plt.barh(p,y,width,color='r',edgecolor='black',label='legend1')
plt.barh(p1,z,width,color='b',edgecolor='black',label='legend2')

plt.yticks(p+width/2,x,rotation=90)
plt.legend()
plt.show()
```

# **Scatter Plot**

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
colors=['r','g','b','y','m','c','k']
sizes=[50,40,30,60,45,70,60]
plt.scatter(day,no,c=colors,s=sizes,alpha=0.5,marker="*",edgecolor='black',linewidth=2)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
colors=[10,49,30,29,56,20,30]
sizes=[100,110,90,60,115,120,130]
plt.scatter(day,no,c=colors,s=sizes,cmap='viridis')
plt.colorbar()
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```

# **Steam Plot**

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,5]
plt.stem(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,5]
plt.stem(x,y,linefmt=":",markerfmt="r+")
plt.show()
plt.stem(x,y,linefmt=":",markerfmt="r_")
plt.show()
plt.stem(x,y,linefmt=":",markerfmt="rd")
plt.show()
```

```python
plt.stem(x,y,linefmt=":",markerfmt="r-")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,5]
plt.stem(x,y,linefmt=":",markerfmt="r+",bottom=2,basefmt="g")
plt.show()
plt.stem(x,y,linefmt=":",markerfmt="r+",bottom=9)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,120]
y=[20,30,40,50,60]
plt.boxplot([x, y])
plt.xticks([1, 2], ['X', 'Y'])
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50,60,70,120]
y = [20,30,40,50,60]

plt.boxplot(
    [x, y],
    showmeans=True,
    flierprops=dict(
        marker='+',
        markerfacecolor='green',
        markeredgecolor='green',
        markersize=10
    ),
    meanprops=dict(
        marker='^',
        markerfacecolor='green'
    ),
    patch_artist=False,
)

plt.xticks([1,2], ['random1', 'random2'])
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50,60,70]
y = [20,30,40,50,60]

plt.boxplot(
    [x, y],
    showmeans=True,
    flierprops=dict(
        marker='+',
        markerfacecolor='green',
        markeredgecolor='green',
        markersize=10
    ),
    meanprops=dict(
        marker='^',
        markerfacecolor='green'
    ),
    boxprops=dict(
        color='red',
    ),
    capprops=dict(
        color='blue',
    ),
    whiskerprops=dict(
        color='green',
    ),
    patch_artist=False,
)

plt.xticks([1,2], ['random1', 'random2'])
plt.show()
```

# **Stack Plot**

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
area1=[2,3,2,5,4]
area2=[2,3,4,5,6]
area3=[1,3,2,4,2]
l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l)
plt.legend()
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[2,3,2,5,4]
area2=[2,3,4,5,6]
area3=[1,3,2,4,2]
l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l,colors=['r','g','b'])
plt.legend()
plt.show()
```

# **Fill_Between Plot**

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,2,3,5,4]
plt.plot(x,y)
plt.fill_between(x,y,color='r')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]
plt.plot(x,y,color='r')
plt.fill_between(x=[2,4],y1=2,y2=4)
plt.show()
```

```python
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])
plt.plot(x,y)
plt.fill_between(x,y,color='r',where=(x>=2) & (x<=4),alpha=0.4)
plt.show()
```

# **Subplot**

```python
import matplotlib.pyplot as plt
import numpy as np
x = [1, 2, 3, 4, 5]
y = [1, 4, 2, 3, 5]
plt.subplot(2, 2, 1)
plt.plot(x, y, color='r', linewidth=2, marker='o')
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)

plt.subplot(2, 2, 2)
categories = ['A', 'B', 'C', 'D', 'E']
values = [3, 7, 2, 5, 8]
plt.bar(categories, values, color='green', edgecolor='black')
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')

plt.subplot(2, 2, 3)
x_scatter = np.random.rand(50) * 10
y_scatter = np.random.rand(50) * 10
plt.scatter(x_scatter, y_scatter, color='blue', alpha=0.7, s=50)
plt.title('Scatter Plot')
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')

plt.subplot(2, 2, 4)
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
colors = ['red', 'green', 'blue', 'yellow']
plt.pie(sizes, labels=labels, colors=colors, autopct='%0.0f%%', startangle=90,radius=1.2)
plt.title('Pie Chart')

plt.tight_layout()
plt.suptitle('4 Different Types of Plots in Matplotlib', fontsize=14, fontweight='bold', y=1.02)
plt.show()
```
